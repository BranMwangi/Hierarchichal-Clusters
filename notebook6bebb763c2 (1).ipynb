```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the csv file.
df1 = pd.read_csv('/kaggle/input/gene-expression/DemoExample_GeneExpression.csv',delim_whitespace=True)
print(df1)
```

```python
sns.clustermap(df1)
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dendrogram=sch.dendrogram(sch.linkage(df1,method='ward'))
```

```python
# Standardize or Normalize every column in the figure
# Standardize:
fig1=sns.clustermap(df1, standard_scale=1)
plt.show()
fig1.savefig('euclidean single.png', bbox_inches='tight')
```

```python
# Normalize
sns.clustermap(df1, z_score=1)
plt.show()
```

```python
#You can use different distance metrics for your data using the metric parameter. The most common methods are correlation and euclidean distance.
# plot with correlation distance
sns.clustermap(df1, metric="correlation", standard_scale=1,cmap="mako")
plt.show()
```

```python
# plot with euclidean distance
sns.clustermap(df1, metric="euclidean", standard_scale=1,cmap="mako")
plt.show()
```

```python
# plot with cosine distance
sns.clustermap(df1, metric="cosine", standard_scale=1,cmap="mako")
plt.show()
```

```python
# linkage method to use for calculating clusters: single
sns.clustermap(df1, metric="euclidean", standard_scale=1, method="single",cmap="viridis")
plt.show()
```

```python
# linkage method to use for calculating clusters: complete
sns.clustermap(df1, metric="euclidean", standard_scale=1, method="complete",cmap="viridis")
plt.show()
```

```python
# linkage method to use for calculating clusters: average
sns.clustermap(df1, metric="euclidean", standard_scale=1, method="average",cmap="viridis")
plt.show()
```
