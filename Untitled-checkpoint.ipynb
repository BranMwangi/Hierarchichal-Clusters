```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read the csv file.
df1 = pd.read_csv(r'C:\Users\Brian\Desktop\DemoExample_GeneExpression.csv',delim_whitespace=True)
print(df1)

```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dendrogram=sch.dendrogram(sch.linkage(df1,method='ward'))
```

```python
hc=AgglomerativeClustering(n_clusters=4, compute_full_tree=True,
    affinity='euclidean', linkage='ward')
hc.fit(df1)
```

```python
from scipy.spatial.distance import pdist
dict_pdist={}
list_metric=[
     'correlation', 'cosine', 'euclidean'
]
for j in list_metric:
    dict_pdist[j] = pdist(X = df1, metric = j)
```

```python
dict_pdist
```

```python
from scipy.cluster.hierarchy import linkage,dendrogram
dict_linkage ={}
list_method = [
    'single','complete','average','weighted'
]
for j in list_metric:
    for k in list_method:
        dict_linkage[j + '_'+  k] = linkage(
            y= dict_pdist[j],
            method = k,
            metric = j,
            optimal_ordering = True
        )
```

```python
dict_linkage['euclidean_complete']
```

```python
#euclidean single linkage dendrogram
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(25,15))
dendrogram(
    Z = dict_linkage['euclidean_single'],
    labels = df1.index,
    count_sort=True,
    distance_sort=True,
    orientation='right',
    leaf_font_size = 15
)
plt.show()
```

```python
#euclidean complete linkage dendrogram
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(25,15))
dendrogram(
    Z = dict_linkage['euclidean_complete'],
    labels = df1.index,
    count_sort=True,
    distance_sort=True,
    orientation='right',
    leaf_font_size = 15
)
plt.show()
```
