```python
# NOW WE HAVE TO AVOID THE SEQS THAT CONTAIN "NC_04"

fragment_to_avoid = ["NC_04"]
counter_max_seqs = 100   # WE WILL PLACE A MAX TO STOP WHEN IT REACHES 100
counter = 1               # AND THIS IS WILL BE THE COUNTER THAT INCREASES VALUE TO REACH THE MAX
for each_id, each_seq in dict_seq.items():
    if counter == counter_max_seqs:
        break
    if each_id[0:5] in fragment_to_avoid:
        continue
    else:
        fragment_to_avoid.append(each_id[0:5])
    with open(f"temp/{each_id}.fasta", "w") as f:
        f.write(f">{each_id}\n{each_seq}")
    genomes.append(f"temp/{each_id}.fasta")
    counter += 1
```

```python
# AND NOW WE RUN THE BIG PART OF THE CODE. WE

for val1, first_genome in enumerate(genomes):
    for val2, second_genome in enumerate(genomes):
        if (first_genome != second_genome) and (ad_array[val1][val2] == 0) :
            p2 = subprocess.run(f"GATEkeeper/bin/GATEkeeper -r {first_genome} -q {second_genome} -o temp/temp", shell=True)
            if p2.returncode == 0:
                mut = 0
                ins = 0
                mut_min_insdels = 0
                with open("temp/temp.vcf", "r") as file_to_read:
                    mut = 0
                    ins = 0
                    dels = 0
                    for line in file_to_read.readlines():
                        if not line.strip().startswith("#") and line:
                            mut += 1
                            if "insert" in line.strip().split("\t")[-1].lower():
                                ins += 1
                            if "delete" in line.strip().split("\t")[-1].lower():
                                dels += 1
                mut_min_insdels = mut - (ins + dels)
                ad_array[val1][val2] = mut
                ad_array[val2][val1] = mut
            else:
                sys.exit(f"error in pair ({first_genome},{second_genome})")
        else:
            print("done")
shutil.rmtree("temp")
            
        
```
